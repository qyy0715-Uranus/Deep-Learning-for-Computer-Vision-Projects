Standard numpy-like indexing and slicing:

```python
tensor = torch.ones(4, 4)
# 创建了一个 $4 \times 4$ 的正方形矩阵 (2D)，里面的每一个元素初始化为 1。
# PyTorch 使用逗号分隔维度：tensor[行, 列]。

print(tensor)

print(f"First row: {tensor[0]}")
# 通过 tensor[0]，我们访问了 tensor 的第一行。由于 tensor 是一个 $4 \times 4$ 的矩阵，第一行包含了四个元素，每个元素的值都是 1。

print(f"First column: {tensor[:, 0]}")
# 冒号 : 表示“所有行”，所以这代表获取第 1 列。

print(f"Last column: {tensor[..., -1]}")
#...（Ellipsis）表示忽略前面的所有维度，-1 表示最后一个。在2D情况下，这等同于获取最后一列。

tensor[:,1] = 0 # 第 2 列的值全部改为 0
print(tensor)
```

Arithmetic Operations 

```python
# 确认tensor初始值
print(tensor)
# This computes the matrix multiplication between two tensors. y1, y2, y3 will have the same value
# ``tensor.T`` returns the transpose of a tensor
y1 = tensor @ tensor.T # python way of doing matrix multiplication
y2 = tensor.matmul(tensor.T) # PyTorch way; matmul() takes one parameter
print (f"y1: \n{y1}\n")
print (f"y2: \n{y2}\n")


#initialization, take the same shape as y1, but fill it with random values

y3 = torch.rand_like(y1)
torch.matmul(tensor, tensor.T, out=y3) #通常函数会返回一个新的结果，但这里我们告诉 PyTorch：“请把计算结果直接写进我们刚才准备好的 y3 空间里，覆盖掉那些随机数。
# 这是一种节省内存的写法，直接将结果写入已经创建好的 y3 中，而不是产生一个新张量。
print (f"y3: \n{y3}\n")
#在深度学习中，频繁地创建和销毁大张量会非常消耗内存和计算资源。通过使用 out= 参数，我们可以复用已有的内存空间，这在处理大规模数据时能显著提高效率。
```

```python
# This computes the element-wise product. z1, z2, z3 will have the same value
z1 = tensor * tensor
z2 = tensor.mul(tensor)

print (f"z1: \n{z1}\n")
print (f"z2: \n{z2}\n")

#cool way of initialition 
z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)

print (f"z3: \n{z3}\n")

# summary: 1. element-wise product: * or mul() 2. matrix multiplication: @ or matmul(), pay attention to the dimension. 
```

```python
# use out= to save memory
A = torch.ones(2, 2)
B = torch.zeros(2, 2) # 初始全是 0
torch.add(A, A, out=B)

# data type: float32 
```

Single-element tensors 

```python
#print(tensor)
print (f"tensor: {tensor}\n")

agg = tensor.sum() #aggregate 

#print(agg, type(agg)) 
print (f"agg: {agg}, type of agg: {type(agg)}\n") #agg is still a tensor, but it has only one element

agg_item = agg.item() #convert a single-element tensor to a Python number, use item() method

#print(agg_item, type(agg_item))
print (f"agg_item: {agg_item}, type of agg_item: {type(agg_item)}\n")
```

In-place operations: Operations that store the result into the operand are called in-place. They are denoted by a _ suffix.

```python
print(f"{tensor} \n")
tensor.add_(5)


print(f"{tensor}")
```

More examples:

```python
x = torch.ones(5)
y = torch.zeros(5)

x
```

```python
x.copy_(y)
# copied the y into x


print (f"x: {x}\n shape: {x.shape}\n")

x.subtract_(5)
print (f"x: {x}\n shape: {x.shape}\n")

# in place operation: the result is stored in the operand, denoted by _ suffix.
```

NOTE:
In-place operations save some memory, but can be problematic when computing derivatives because of an immediate lossof history. Hence, their use is discouraged.

Tensor to NumPy array

```python
t = torch.ones(5)
print(f"t: {t}")


n = t.numpy() #use numpy()
print(f"n: {n}")
```

Numpy array to (torch) tensor 

```python
import numpy as np

n = np.ones(5)

print(f"n: {n}\n shape: {n.shape}\n")


t = torch.from_numpy(n)
print(f"t: {t} \n shape: {t.shape}\n")
```
